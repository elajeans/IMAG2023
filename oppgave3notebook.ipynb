# Oppgave 3: Poissonligning (1D)

## Problemstilling

Vi skal løse Poissonligningen

$$
u_{xx}(x) = \cos(\pi x), \qquad -1 \le x \le 1,
$$

med Dirichlet-randbetingelser

$$
u(-1) = 0, \qquad u(1) = 2.
$$


## (a) Analytisk løsning

Ligningen integreres to ganger med hensyn på $x$:

$$
u_x(x) = \frac{1}{\pi}\sin(\pi x) + C_1
$$

$$
u(x) = -\frac{\cos(\pi x)}{\pi^2} + C_1 x + C_2
$$

Randbetingelsene gir systemet (merk at $\cos(-\pi)=\cos(\pi)=-1$):

$$
u(-1)=0: \quad \frac{1}{\pi^2} - C_1 + C_2 = 0
$$

$$
u(1)=2: \quad \frac{1}{\pi^2} + C_1 + C_2 = 2
$$

Addisjon og subtraksjon av likningene gir

$$
C_1 = 1, \qquad C_2 = 1 - \frac{1}{\pi^2}.
$$

Den analytiske løsningen blir derfor

$$
u(x) = -\frac{\cos(\pi x)}{\pi^2} + x + 1 - \frac{1}{\pi^2}.
$$


## (b) Numerisk løsning

### Diskretisering

Intervallet $[-1,1]$ diskretiseres med $m=4$ indre punkter.  
Steglengden blir

$$
h = \frac{2}{m+1} = \frac{2}{5} = 0.4.
$$

Andrederiverten approksimeres med sentral differanse

$$
u_{xx}(x_i) \approx
\frac{u_{i-1} - 2u_i + u_{i+1}}{h^2}, \qquad i=1,\dots,m.
$$

Dette gir et lineært system

$$
L\mathbf{u} = \mathbf{b}.
$$

Dirichlet-randbetingelsene bygges inn i høyresiden:

$$
b_1 \leftarrow b_1 - \frac{u(-1)}{h^2}, \qquad
b_m \leftarrow b_m - \frac{u(1)}{h^2}.
$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Antall indre punkter
m = 4

# Romgitter
x = np.linspace(-1, 1, m + 2)
h = x[1] - x[0]

# Laplace-matrise
L = (1 / h**2) * (
    np.diag((m-1) * [1], -1) +
    np.diag(m * [-2], 0) +
    np.diag((m-1) * [1], 1)
)

# Høyreside
f = np.cos(np.pi * x[1:-1])
b = f.copy()
b[0] -= 0 / h**2
b[-1] -= 2 / h**2

# Løs systemet
u_inner = np.linalg.solve(L, b)

# Sett sammen full løsning
u = np.concatenate([[0], u_inner, [2]])
u
```

### Analytisk løsning


```python
def u_exact(x):
    return -np.cos(np.pi*x)/np.pi**2 + x + 1 - 1/np.pi**2

x_fine = np.linspace(-1, 1, 200)
error = np.max(np.abs(u - u_exact(x)))
error
```

```python
plt.figure(figsize=(6,4))
plt.plot(x_fine, u_exact(x_fine), label='Analytisk løsning')
plt.plot(x, u, 'o--', label='Numerisk løsning (m=4)')
plt.xlabel('x')
plt.ylabel('u(x)')
plt.legend()
plt.grid(True)
plt.show()
```

## Resultat

Den maksimale feilen mellom numerisk og analytisk løsning er

$$
\max_i |u_i - u(x_i)|.
$$

Selv med et grovt gitter ($m=4$) gir metoden en god tilnærming til den analytiske løsningen.
