# Oppgave 5: Laplacelikning (2D)
## Problemstilling
Vi skal løse Laplacelikningen
$$u_{xx}(x,y) + u_{yy}(x,y) = 0$$
for området
$$-5 < x < 5, \quad 0 < y < 2$$
med randbetingelser
$$u(-5,y)=\sin(2\pi y), \quad u(5,y)=\sin(2\pi y)$$
$$u(x,0)=0, \quad u(x,2)=\sin(\pi x)$$
Laplacelikningen beskriver stasjonær varmefordeling i et rektangulært domene.


## Diskretisering
Domenet diskretiseres med $N_x = 41$ punkter i x-retning og $N_y = 21$ punkter i y-retning.
Steglengdene blir
$$\Delta x = \frac{10}{N_x-1} = 0.25$$
$$\Delta y = \frac{2}{N_y-1} = 0.1$$

Andrederiverte approksimeres med sentral differanse:
$$u_{xx}(x_i,y_j) \approx \frac{u_{i-1,j}-2u_{i,j}+u_{i+1,j}}{\Delta x^2}$$
$$u_{yy}(x_i,y_j) \approx \frac{u_{i,j-1}-2u_{i,j}+u_{i,j+1}}{\Delta y^2}$$

Dersom $\Delta x = \Delta y = h$ får vi fempunkts-approksimasjonen
$$u_{i,j} = \frac14 (u_{i+1,j}+u_{i-1,j}+u_{i,j+1}+u_{i,j-1})$$
som betyr at hvert indre punkt er gjennomsnittet av sine fire naboer.


## Numerisk løsning – Jacobi-metoden
Ligningssystemet løses iterativt med Jacobi-metoden.
Ved hvert steg oppdateres alle indre punkter basert på verdiene fra forrige iterasjon.
Randbetingelsene holdes faste under hele prosessen.


```python
import numpy as np
import matplotlib.pyplot as plt

Nx = 41
Ny = 21

x = np.linspace(-5,5,Nx)
y = np.linspace(0,2,Ny)

dx = x[1]-x[0]
dy = y[1]-y[0]

u = np.zeros((Ny,Nx))

# randbetingelser
for j in range(Ny):
    u[j,0] = np.sin(2*np.pi*y[j])
    u[j,-1] = np.sin(2*np.pi*y[j])

for i in range(Nx):
    u[0,i] = 0
    u[-1,i] = np.sin(np.pi*x[i])

iterations = 5000

for k in range(iterations):
    u_new = u.copy()
    u_new[1:-1,1:-1] = 0.25*(
        u[1:-1,2:] +
        u[1:-1,:-2] +
        u[2:,1:-1] +
        u[:-2,1:-1]
    )
    u[1:-1,1:-1] = u_new[1:-1,1:-1]
```

```python
X,Y = np.meshgrid(x,y)

plt.figure(figsize=(7,4))
plt.contourf(X,Y,u,50)
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Numerisk løsning av Laplacelikningen')
plt.show()
```

## Resultat
Løsningen etter 5000 Jacobi-iterasjoner viser temperaturfordelingen i domenet.
Randbetingelsene styrer løsningen: sinusvariasjon langs sidene,
null langs bunnen og sinus langs toppen.
